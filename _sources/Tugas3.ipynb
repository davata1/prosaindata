```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import random

nltk.download('punkt')
nltk.download('stopwords')

# Fungsi untuk menghapus tanda baca
def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

# Fungsi untuk menghapus stop words
stop_words = set(stopwords.words('english'))
def remove_stopwords(text):
    words = word_tokenize(text)
    filtered_words = [word for word in words if word.lower() not in stop_words]
    return ' '.join(filtered_words)

# Fungsi untuk melakukan stemming
stemmer = PorterStemmer()
def stem_text(text):
    words = word_tokenize(text)
    stemmed_words = [stemmer.stem(word) for word in words]
    return ' '.join(stemmed_words)

# Fungsi untuk normalisasi teks abstrak
def normalize_abstract(text):
    text = remove_punctuation(text)
    text = remove_stopwords(text)
    text = stem_text(text)
    text = text.lower()
    text = text.strip()

    return text

data = []

for i in range(1, 30):
    url = f"https://pta.trunojoyo.ac.id/welcome/index/{i}"  # Sesuaikan URL dengan halaman yang sesuai
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response, 'html.parser')

    lis = soup.find_all('li', {'data-id': 'id-1'})  # Menggunakan atribut 'data-id'

    for li in lis:
        judul = li.find('a', class_='title').text  # Menemukan teks dalam elemen <a> dengan kelas 'title'
        deskripsi = li.find('span').text
        abstrak = li.find('p').text
        normalized_abstrak = normalize_abstract(abstrak)  # Normalisasi teks abstrak
        tokens = word_tokenize(abstrak)  # Tokenisasi teks abstrak

        # Menambahkan label (misalnya, label positif 1 dan label negatif -1)
        label = 1 if random.randint(0, 1) == 1 else -1

        data.append([judul, deskripsi, abstrak, normalized_abstrak, tokens, label])

# Membuat DataFrame dengan tambahan kolom 'Judul', 'Deskripsi', 'Abstrak_Normalized', 'Tokens', dan 'Label'
df = pd.DataFrame(data, columns=['Judul', 'Deskripsi', 'Abstrak', 'Abstrak_Normalized', 'Tokens', 'Label'])
print(df)

df.to_csv('data_berita.csv', index=False)
```

```python
# Import library yang diperlukan
from nltk.tokenize import word_tokenize

# Fungsi untuk mengubah teks abstrak menjadi Term Frequency
def abstract_to_tf(text):
    # Tokenisasi teks
    tokens = word_tokenize(text)

    # Jika abstrak kosong atau tidak ada kata, kembalikan nilai default 0
    if len(tokens) == 0:
        return {'N/A': 0}

    # Hitung frekuensi kemunculan setiap kata dalam teks
    word_count = {}
    for word in tokens:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    # Hitung TF untuk setiap kata
    tf = {}
    total_words = len(tokens)
    for word, count in word_count.items():
        tf[word] = count / total_words

    return tf

# Assuming you have a column 'Abstrak' containing abstract text
df['Abstrak_TF'] = df['Abstrak'].apply(abstract_to_tf)

# Then, add the 'Abstrak_TF_Values' column
df['Abstrak_TF_Values'] = df['Abstrak_TF'].apply(lambda x: list(x.values()) if x != {'N/A': 0} else [0])

# Finally, display the DataFrame
print(df[['Judul', 'Abstrak_TF_Values']])

```

```python
# Import library yang diperlukan
from nltk.tokenize import word_tokenize

# Fungsi untuk mengubah teks abstrak menjadi Term Frequency
def abstract_to_tf(text):
    # Tokenisasi teks
    tokens = word_tokenize(text)

    # Jika abstrak kosong atau tidak ada kata, kembalikan nilai default 0
    if len(tokens) == 0:
        return {'N/A': 0}

    # Hitung frekuensi kemunculan setiap kata dalam teks
    word_count = {}
    for word in tokens:
        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1

    # Hitung TF untuk setiap kata
    tf = {}
    total_words = len(tokens)
    for word, count in word_count.items():
        tf[word] = count / total_words

    return tf

# Mengganti nilai-nilai dalam TF menjadi Binary Frequency (1 jika lebih dari 0, 0 jika 0)
df['Abstrak_Binary_Frequency'] = df['Abstrak_TF'].apply(lambda x: [1 if val > 0 else 0 for val in x.values()])

# Menampilkan DataFrame dengan kolom Binary Frequency
print(df[['Judul', 'Abstrak_Binary_Frequency']])

```

# Crawling Twitter


```python
import snscrape.modules .twitter as sntwitter
import pandas as pd

data_tweet= []

#mengambil data rentang waktu tertentu 
for data, tweet in enumerate(sntwitter.TwitterSearchScraper("Cak imin since:2022-01-01 until:2022-02-22").get_items()):
  if data >=300:
    break
  data_tweet.append([tweet.user.username,tweet.content,tweet.user.location])
```

```python
hasil_data_twitter =pd.DataFrame(data_tweet, columns=["Akun","tweet","Lokasi"])
hasil_data_twitter
```

```python
hasil_data_twitter.to_csv("Twitter.csv")
```

## Data PTA

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd 
import csv

#permintaan dari web browser ke web server untuk menampilkan sesuai dengan SO yang digunakan
server= {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.5481.104 Safari/537.36"}
```

```python
def scrape_TA(page):
  global server
  data=[]
  for laman in range(1,page+1):
    if laman==1 :
      URL = "https://pta.trunojoyo.ac.id/c_search/byprod/10/"
    else :
      URL = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{laman}"
    request = req.get(URL,server).text
    #var bs yang menyimpan data request berupa html
    soup = bs(request, 'lxml')
    prodi = soup.find_all('div',{'id':'begin'})
    for pro in prodi:
      prod = pro.find('h2').text
    jurusan = prod[-18:]
    ul = soup.find('ul', 'items list_style')
    li = ul.find_all('li', {'data-id':'id-1'})
    for x in li:
      link = x.find('a','gray button')['href']
      request2 = req.get(link, server).text
      soup2 = bs(request2, 'lxml')
      abstrak= soup2.find('p',{'align':'justify'}).text.replace('\r','').replace('\n','')
      NPM = x.find('a','gray button')['href'][-12:]
      headline = x.find('a', 'title').text.replace('\r','').replace('\n','')
      data.append([NPM,headline,abstrak,jurusan])
  return data
```

```python
data_TA= scrape_TA(20)
data_TA_result=pd.DataFrame(data_TA)
data_TA_result.columns = ["NPM", "Judul","Abstrak","jurusan"]
data_TA_result
```

```python
data_TA_result.to_csv("Data_PTA.csv")
```

## Berita Online

```python
import requests as req
from bs4 import BeautifulSoup as bs
import pandas as pd 
import csv

#permintaan dari web browser ke web server untuk menampilkan sesuai dengan SO yang digunakan
server= {"user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.5481.104 Safari/537.36"}
```

```python
def scrape_detik (hal):
  global server
  data=[]
  for page in range (1, hal+1):
    #print(page)
    # mengambil data banjir kemudian diurutkan berdasarkan waktu dan halaman
    url = f"https://www.detik.com/search/searchall?query=hujan&siteid=2&sortby=time&page={hal}"
    request = req.get(url,server).text
    # memanggil beautysoup
    soup = bs(request, "lxml")
    #menampilkan list berita
    daftar_berita = soup.find("div","list media_rows list-berita")
    # menampil isi artikel
    artikel = daftar_berita.find_all("article") 

    for x in artikel:
      #judul
      judul = x.find_all("h2")
      judul = [ele.text.strip() for ele in judul]
      # deskripsi
      isi  = x.find_all("p")
      isi  = [ele.text.strip() for ele in isi]
      data.append([judul[0], isi[0]])
  return data
```

```python
berita=scrape_detik(20)
isi_berita = pd.DataFrame(berita)
isi_berita.columns=["Judul","Isi berita"]
isi_berita
```
